```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
fsr=1000
omega=25
a = np.linspace(0, 10, 10000)
r=0.99995

f_omega = r* ( np.exp(1j*a)-1   )/(1-(r**2) *np.exp(1j*a)  )

f_plus = r* ( np.exp(1j*a + 1j * (omega/fsr)  )-1  )/ (1-(r**2) *np.exp(1j*a + 1j * (omega/fsr)  ) )

f_minus = r* ( np.exp(1j*a - 1j * (omega/fsr)  )-1  )/(1-(r**2) *np.exp(1j*a - 1j * (omega/fsr)  )   )                                                   


```

```python
y=(f_omega * np.conj(f_plus) -np.conj(f_omega) * f_minus) * -1
y_imag=y.imag
```

```python
# Plotting the real and imaginary parts
plt.figure(figsize=(10, 5))
plt.plot(a, y_imag, label='Error signal')


plt.xlim(6.2, 6.4)  # Set x-axis range from 2 to 8
#plt.title('Complex Exponential: $e^{i\\omega t}$')
plt.xlabel('Free spectral range')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Plotting the real and imaginary parts
plt.figure(figsize=(10, 5))
plt.plot(a, f_omega.real , label='Error signal')


plt.xlim(6,6.5)  # Set x-axis range from 2 to 8
#plt.title('Complex Exponential: $e^{i\\omega t}$')
plt.xlabel('Time (t)')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
F = 12  # finesse
delta = np.linspace(0.5, 2.5, 1000)  # normalized detuning

# Transmission formula (Airy function)
T = 1 / (1 + (4 * F**2 / np.pi**2) * np.sin(np.pi * delta)**2)

# Plot
plt.plot(delta, T,label='Transmission Intensity')
plt.xlim(0.5, 2.5) 
#plt.title("Transmission of a FabryPerot Cavity")
plt.xlabel("Free spectral range")
plt.ylabel("Transmission")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
R = 0.99995  # mirror reflectivity
#F = np.pi * np.sqrt(R) / (1 - R)  # finesse
F=12
delta = np.linspace(-0.5, 0.5, 1000)  # normalized detuning

# Reflected intensity formula
I_reflected = ((1 - R)**2 + 4 * R * np.sin(np.pi * delta)**2) / (1 + (4 * F**2 / np.pi**2) * np.sin(np.pi * delta)**2)

# Plot
plt.plot(delta, I_reflected,label='Reflected Intensity')
plt.xlim(-0.5, 0.5) 
#plt.title("Reflected Intensity from FabryPerot Cavity")
plt.xlabel("Free spectral range")
plt.ylabel("Reflected Intensity")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Create subplots side by side
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# First plot
axs[0].plot(delta, I_reflected,label='Reflected Intensity')
#axs[0].set_xlim(-6, 6.5)
axs[0].set_xlabel("Free spectral range")
axs[0].set_ylabel('Amplitude')
axs[0].legend()
axs[0].grid(True)


# Second plot
axs[1].plot(a, y_imag, label='Error signal')
axs[1].set_xlim(6.2, 6.4)
axs[1].set_xlabel("Free spectral range")
axs[1].set_ylabel('Amplitude')
axs[1].legend()
axs[1].grid(True)

```

```python
time=50*(10**-3)
x=np.linspace(0,time,500)

f0=3*10**14
phi0=1
fm=25*10**6

def sideband(t):
    y= np.sin( 2*np.pi*f0*t + phi0* np.sin(2*np.pi*(f0+fm)*t)  )
    
    return y

# Plot
plt.plot(x, sideband(x),label='Sideband')

#plt.title("Reflected Intensity from FabryPerot Cavity")
plt.xlabel("Time")
plt.ylabel("Reflected Intensity")
plt.legend()
plt.grid(True)
plt.show()

```
