```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df1=pd.read_csv('daiyan_trace_1.csv')
df2=pd.read_csv('daiyan_trace_2.csv')
df2.head()
```

```python
freq1=df1['Freq(Hz)']
volt1=df1['Voltage(dBm)']

freq2=df2['Frequency(Hz)']
volt2=df2['Voltage(dBm)']
```

```python
# Plot using Matplotlib
plt.plot(freq1,volt1, color='blue',label="Plot 2")
plt.plot(freq2,volt2, color='red',label='Plot 1')

# Labels and title
plt.xlabel('Frequency(Hz)')
plt.ylabel('Voltage(dBm)')
plt.legend()
# Show grid and plot
plt.grid(True)
plt.show()
```
